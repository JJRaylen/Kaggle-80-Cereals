## Project: Nutrition Data Visualisation on 80 cereal products
Link: https://www.kaggle.com/crawford/80-cereals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df = pd.read_csv('./Nutrition/cereal.csv')
```

```python
fields = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis=1)
cereal_df_new.head()
```

```python
cereal_corr = cereal_df_new.corr() #Get correlation data
cereal_corr
```

```python
A = np.ones_like(cereal_corr, dtype=bool)
mask = np.triu(A)
mask
```

```python
cereal_corr_new = cereal_corr.iloc[1:, :-1]
mask_new = mask[1:, :-1]
mask_new
```

```python
fig, ax = plt.subplots(figsize=(12,9))
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

ax = sns.heatmap(data=cereal_corr_new, mask=mask_new, annot=True, 
                annot_kws={"fontsize":13}, fmt=".2f", 
                vmin=-1, vmax=1,
                linecolor = "white", linewidths=2, cmap="YlGnBu", cbar=True);
yticks = [i.upper() for i in cereal_corr_new.index]
xticks = [i.upper() for i in cereal_corr_new.columns]

ax.set_yticklabels(yticks,rotation=0, fontsize=13);
ax.set_xticklabels(xticks,rotation=0, fontsize=13);

title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'
ax.set_title(title, loc='left', fontsize=18);
```
